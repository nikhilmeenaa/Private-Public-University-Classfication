```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200 , n_features=2 , centers = 5)
```

```python
data[0]   # => data[0] is the actual data generated

# => data[1], while this is the cluster (label), to which that particular data point belongs
```

```python
plt.scatter( data[0][:,0:1] , data[0][:,1:] ,c = data[1] , cmap = "rainbow")
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (org, clustered) = plt.subplots(1 , 2 , figsize = (10,5))
org.scatter(data[0][:,0:1] , data[0][:, 1:] , c = data[1] , cmap = "rainbow")
clustered.scatter(data[0][:,0:1] , data[0][:, 1:] , c = kmeans.labels_ , cmap = "rainbow")
```
